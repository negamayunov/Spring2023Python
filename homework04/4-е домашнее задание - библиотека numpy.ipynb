6.  Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def nxn(n):
    a = np.arange(n)
    b = a.copy()
    a = a[:, np.newaxis]
    return a + b

n = int(input('n = '))
nxn(n)
```
